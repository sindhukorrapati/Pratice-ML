```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/emails.csv")
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
X = data.iloc[:,1:3001]
```

```python
X
```

```python
Y = data.iloc[:,-1]
```

```python
Y.values
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.22)
```

```python
NB = MultinomialNB(alpha=1.9)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
NB.fit(x_train,y_train)
```

```python
NB_ypredict1 = NB.predict(x_test)
```

```python
NB_ypredict1
```

```python
print("Accuracy of Navie Bayes Classifier Model is {}".format(accuracy_score(y_test,NB_ypredict1)))
```

```python
RF = RandomForestClassifier(n_estimators=100,criterion="gini")
```

```python
RF.fit(x_train,y_train)
```

```python
RF_ypredict2 = RF.predict(x_test)
```

```python
RF_ypredict2
```

```python
print("Accuracy of Random Forest Classifier model is {}".format(accuracy_score(y_test,RF_ypredict2)))
```
